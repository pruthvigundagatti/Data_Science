#**Association Rule Mining**

**Apriori algorithm**

```python
df_encoded=df['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].str.get_dummies(sep=',')
df_encoded
```

```python
#Apply apriori to the one-hot encoded DataFrame
frequent_itemsets=apriori(df_encoded,min_support=0.01,use_colnames=True)
frequent_itemsets
```

#Generate Association Rule

```python
rules = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.01)
rules
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=0.01)
rules
```

#**Analysis and Interpretation**

#Filter rules with high confidence and lift

```python
high_confidence_rules=rules[(rules['confidence']>=0.1) & (rules['lift']>1.0)]
high_confidence_rules
```

```python
#Sort by lift to find the strongest rules
sorted_rules=high_confidence_rules.sort_values(by='lift',ascending=False)
sorted_rules
```

#**Interview Question:**

# 1.	What is lift and why is it important in Association rules?

Lift is a measure that quantifies the strength of an association rule compared to random chance. It is calculated as the ratio of the observed support to what would be expected if the two itemsets were independent.

Lift > 1: A and B occur more frequently together than would be expected by chance (positive correlation)

Lift = 1: A and B occur together as expected by chance (no correlation).

Lift < 1: A and B occur together less frequently than expected (negative correlation).

# **Importance:**
Lift helps determine how useful a rule is in predicting one item based on another. It provides insights into whether the co-occurrence of items is just random or truly significant.

# 2.	What is support and Confidence. How do you calculate them?

# **Support:**
It indicates how frequently an itemset appears in the dataset. Support measures the proportion of transactions that contain a particular itemset.

Support(A)= Transactions containing A / Total number of transactions

# **Confidence:**
Confidence measures how often items in A appear in transactions that also contain 𝐵. It is the conditional probability of buying B given that
A has been bought.

Confidence(A→B)= Support(A∩B) / Support(A)

# 3. What Are Some Limitations or Challenges of Association Rule Mining?

1. Scalability: Association rule mining can be computationally expensive when dealing with large datasets, as the number of possible item combinations grows exponentially.

2. Too Many Rules: If not carefully managed (e.g., by setting appropriate thresholds for support and confidence), mining can generate a large number of rules, many of which may be trivial or uninteresting.

3. Low Support for Rare Events: Some potentially interesting but rare patterns might not meet the minimum support threshold, leading to their exclusion from the analysis.

4. Interpretability: The sheer number of rules can sometimes overwhelm the analyst, making it challenging to focus on the most actionable insights.
