#**Interview Questions:**

1. Can you explain the difference between user-based and item-based collaborative filtering?


User-Based Collaborative Filtering (UBCF): Recommends items based on the preferences of similar users. It compares users to find those with similar tastes and suggests items they've liked.


Item-Based Collaborative Filtering (IBCF): Recommends items similar to those the user has previously interacted with. It compares items to suggest others that are often liked by users who liked the same items.

2. What is collaborative filtering, and how does it work?

Collaborative filtering is a recommendation system technique that predicts a user's preferences by analyzing the preferences of many users. It works by finding patterns in user interactions (like ratings or purchases) with items to suggest new items the user might like.

User-Based: Recommends items based on what similar users have liked.
Item-Based: Recommends items that are similar to what the user has previously liked.
