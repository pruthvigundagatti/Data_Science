# **Data Preprocessing:**

# Impute missing values.

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data['Fare'].fillna(train_data['Fare'].median(), inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
test_data.isnull().sum()
```

# Encoding categorical variables using one-hot encoding

```python
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
train_data
```

```python
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
test_data
```

# Drop irrelevant columns

```python
train_data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
train_data
```

```python
test_data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
test_data
```

# **Split Data for Training and Testing**

```python
from sklearn.model_selection import train_test_split
```

```python
x=train_data.drop('Survived',axis=1)
y=train_data['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Build Predictive Models (LightGBM and XGBoost)

```python
# First import the LGBM and XGBM

import lightgbm as lgb
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

# LightGBM model

```python
lgbm_model = lgb.LGBMClassifier(random_state=42)
lgbm_model.fit(x_train, y_train)
```

```python
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
y_pred_lgbm = lgbm_model.predict(x_test)
```

```python
y_pred_xgb = xgb_model.predict(x_test)
```

# LGBM accuracy precision recall and f1_score

```python
accuracy_lgb = accuracy_score(y_test, y_pred_lgbm)
print(f"LightGBM Accuracy: {accuracy_lgb}")
```

```python
precision_lgb = precision_score(y_test, y_pred_lgbm)
print(f"LightGBM Precision: {precision_lgb}")
```

```python
recall_lgb = recall_score(y_test, y_pred_lgbm)
print(f"LightGBM Recall: {recall_lgb}")
```

```python
f1_lgb = f1_score(y_test, y_pred_lgbm)
print(f"LightGBM F1 Score: {f1_lgb}")
```

# XGBM accuracy precision recall and f1_score

```python
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb}")
```

```python
precision_xgb = precision_score(y_test, y_pred_xgb)
print(f"XGBoost Precision: {precision_xgb}")
```

```python
recall_xgb = recall_score(y_test, y_pred_xgb)
print(f"XGBoost Recall: {recall_xgb}")
```

```python
f1_xgb = f1_score(y_test, y_pred_xgb)
print(f"XGBoost F1 Score: {f1_xgb}")
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Tuning parameters for LightBM

lgb_params_grid= {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7]
}
lgbm_grid = GridSearchCV(lgb.LGBMClassifier(random_state=42), lgb_params_grid, cv=5, scoring='accuracy')
lgbm_grid.fit(x_train, y_train)
print(f"Best parameters for LightGBM: {lgbm_grid.best_params_}")
```

```python
xgb_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7]
}

xgb_grid = GridSearchCV(xgb.XGBClassifier(random_state=42), xgb_param_grid, cv=5, scoring='accuracy')
xgb_grid.fit(x_train, y_train)
print(f"Best parameters for XGBoost: {xgb_grid.best_params_}")
```

# Comparative Analysis

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
lgb_scores = [accuracy_lgb, precision_lgb, recall_lgb, f1_lgb]
xgb_scores = [accuracy_xgb, precision_xgb, recall_xgb, f1_xgb]
```

```python
# Bar plot for comparison
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - width/2, lgb_scores, width, label='LightGBM')
ax.bar(x + width/2, xgb_scores, width, label='XGBoost')

ax.set_xlabel('Metrics')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

plt.show()
```
