# **Data Preprocessing**

Encode the categorical variables if necessary

```python
from sklearn.preprocessing import LabelEncoder
```

Encode all categorical features

```python
label_encoders={}
for column in df.columns:
  le=LabelEncoder()
  df[column]=le.fit_transform(df[column])
  label_encoders[column]=le
```

```python
df.head()
```

Split the dataset into training and testing data sets

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('class',axis=1)
y=df['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# **SVM implementation**

Implement a basic SVM classifier using scikit-learn

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

Initialize the SVM classifier

```python
svm_model=SVC()
svm_model.fit(x_train,y_train)
```

```python
y_pred=svm_model.predict(x_test)
```

evaluate the performance

```python
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_test, y_pred))
```

# **Visualization of SVM Results**

visualize classification results on the testing data, this can be challenging with high dimensional data, but we can use PCA to reduce the dimensionality and visualize the classification

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
x_test_pca=pca.fit_transform(x_test)
```

```python
plt.scatter(x_test_pca[:, 0], x_test_pca[:, 1], c=y_pred, cmap='coolwarm')
plt.title('SVM Classification Results')
plt.show()
```

Experiment with different SVM hyperparameters(kernel,C,gamma,etc.)

```python
svm_rbf=SVC(kernel='rbf',C=1.0,gamma='scale')
svm_rbf.fit(x_train,y_train)
y_pred_rbf=svm_rbf.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
```
